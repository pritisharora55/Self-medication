# Calculation of BMI with Height and Weight

This is used to convert Height in feet to the nearest integer value, so I have used float here. 

This is done because for BMI, height is required in meters and at the point of collection of data, I didnt think that I would calculate BMI so the forms had height in Feet and Inches wheresa for BMI it is needed in Meters.

```python
final['Floor'] = final['Height'].apply(np.floor)
```

```python
final['Nheight'] = final['Height']-final['Floor']
```

```python
final["Heightf"] = ((final['Floor'] *.3048) +( final['Nheight'] * .0254))
```

```python
final.head()
```

```python
final['Height^2'] = final['Heightf']** 2
```

```python
final.head()
```

BMI is calculated here and then added as a dataform.

```python
final['BMI'] = final["Weight"]/final["Height^2"]
```

```python
final.head()
```

```python
final = final.drop(['Height',"Weight","Floor",'Nheight','Height^2','Heightf'],axis = 'columns')
```

```python
final.head()
```

Here 1 means Yes and 0 means No


1 Means Male and 0 means Female.


As string values don't mean much for models, they are required to be changed into numeric values so they can be implemented in models. So, I have used Encoder function in Scikit Learn to change string values to numeric.

```python
from sklearn.preprocessing import LabelEncoder
Sex = LabelEncoder()
Smoker = LabelEncoder()
Diabetic = LabelEncoder()
Alcohol = LabelEncoder()
BP = LabelEncoder()
Fever = LabelEncoder()
Migraine = LabelEncoder()
Cold = LabelEncoder()
Digestion = LabelEncoder()
final['Sex'] = Sex.fit_transform(final['Gender'])
final['Smoker ?'] = Smoker.fit_transform(final['Do you smoke ?'])
final['Diabetic ?'] = Diabetic.fit_transform(final['Are you Diabetic?'])
final['Alcohol Consumption ?'] = Alcohol.fit_transform(final['Do you consume alcohol?'])
final['Blood Pressure ?'] = BP.fit_transform(final['Do you take any medication for High or Low Blood Pressure?'])
final['Did self medication work for fever?'] = Fever.fit_transform(final['Self Medication helpful in fever ?'])
final['Did self medication work for migraine/Headache?'] = Migraine.fit_transform(final['Self Medication helpful in Migraine ?'])
final['Did self medication work for cold?'] = Cold.fit_transform(final['Was Self Medication helpful in Cold/Cough'])
final['Did self medication work for digestive problems?'] = Digestion.fit_transform(final['Was self medication effective in Digestive Problems?'])
```

```python
final.head()
```

```python
final = final.drop(['Gender','Do you smoke ?','Are you Diabetic?','Do you consume alcohol?','Do you take any medication for High or Low Blood Pressure?','Self Medication helpful in fever ?','Self Medication helpful in Migraine ?','Was Self Medication helpful in Cold/Cough','Was self medication effective in Digestive Problems?'], axis = 'columns')
```

# Selecting datasets for training

```python
X = final.drop(['Overall Self Medication Helpful = Total Yes/Total Answers(Yes + No)','Blood Pressure ?'], axis='columns')
X
```

```python
y = final['Overall Self Medication Helpful = Total Yes/Total Answers(Yes + No)']
```

# Model Selection

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size = 0.25)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion = 'entropy')
model.fit(X_train, y_train)
```

```python
predicitions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
m1 = accuracy_score(y_test,predicitions) * 100
m1
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 50, criterion ='entropy', min_samples_split = 4)
model.fit(X_train, y_train)
```

```python
m2 = model.score(X_test,y_test) * 100
m2
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
m3 = model.score(X_test,y_test) * 100
m3
```

```python
data = [['Decision Tree',m1],['Random Forest',m2],['Logistic Regression',m3]]
df = pd.DataFrame(data, columns = ['Model','Accuracy'])
```
